```python
content=open('/content/rosalind_fibd.txt')
f=content.read().split(' ')
n=int(f[0])
m=int(f[1])
content.close()

rabbits = [1, 1]  #rabbits become fertile at 1 month old so the reproduction begins on the second month.                                                             
months = 2 #months are equal to the total number of items in the list even when they increase later on in the code
                                                                    
while months < n:                                                              
    if months < m:  #if your month is below the month they begin to die                                                         
        rabbits.append(rabbits[-2] + rabbits[-1])   #apply N_current=N_one_month_back + N_twomontsh_back 

    elif months == m:  #if your month is equal to the month they begin to die                                      
        rabbits.append(rabbits[-2] + rabbits[-1] - rabbits[-m])  #apply above less the rabbits in the first month )                      
#(the first month is [-m]"minus" because if you use a minus sign when indexing it indexes from last to first) 
# and "m" because if they have started dying then it means that the months that has passed are equal to m]   
    else:  #if your month is NOT below NOR equal to the month they begin to die then it's above and they continue to die                                                                   
        rabbits.append(rabbits[-2] + rabbits[-1] - rabbits[-(m + 1)])  #same as above only now you use [-(m+1)]                  
#because you want to indicate the rabbits dying are the ones born the next month from the ones that died previously
    months += 1 #after any conditional is satisfied increase month by 1 to keep going until n                                                              
print(rabbits[-1]) #print the last month which is the last entry in the list going left to right




```
